Notebook này khởi tạo và chạy pipeline để extract semantic vector (SBERT) cho Content-Based Filtering thông qua file Parquet, và ghi ra file `memmap` như kế hoạch thực thi.

```python
import sys
import os
import numpy as np
import pandas as pd

# Add scripts to sys.path
%pip install faiss-cpu
BASEDIR = '/content/drive/My Drive/Thesis/book_recsys'
sys.path.append(os.path.abspath(BASEDIR))
from scripts.item_vector_builder import ItemVectorBuilder
from scripts.faiss_builder import FaissBuilder
```

```python
# Định nghĩa constants
VERSION = "v1"
_PROC              = f'{BASEDIR}/data/processed_2/'
INPUT_PQ           = f'{_PROC}books_cleaned.parquet'#f'{_PROC}books_filtered.parquet'

_OUTDIR            = f'{_PROC}model_{VERSION}/'
os.makedirs(_OUTDIR, exist_ok=True)

OUTPUT_MATRIX  = f'{_OUTDIR}cb_sbert_matrix_v2.npy'
OUTPUT_INDEX       = f'{_OUTDIR}cb_sbert_book_index.csv'
OUTPUT_CKPT        = f'{_OUTDIR}cb_sbert_progress.json'
MODEL_NAME = 'sentence-transformers/all-MiniLM-L6-v2'


CHUNK_SIZE = 50000
ENCODE_BATCH = 256
MAX_DESC_CHARS = 300

builder = ItemVectorBuilder(
    input_parquet=INPUT_PQ,
    output_matrix_path=OUTPUT_MATRIX,
    output_index_path=OUTPUT_INDEX,
    model_name=MODEL_NAME,
    chunk_size=CHUNK_SIZE,
    encode_batch_size=ENCODE_BATCH,
    max_desc_chars=MAX_DESC_CHARS,
    device=None
)
```

```python
# Kích hoạt tiến trình map vectors
builder.build()
```

```python
# --- CELL BỔ SUNG: VÉT CỨU HỘ HOẶC BUILD ĐỘC LẬP FILE INDEX CSV ---
# Chạy cell này sẽ sử dụng hàm độc lập build_index_csv() ta vừa xây trong Class.
# Nó đọc trực tiếp từ Parquet theo đúng thứ tự mảng gốc rồi dump chốt 1 lần xuống Drive.

try:
    print("Tiến hành chạy hàm vét CSV Index từ Parquet gốc...")
    id_map_df = builder.build_index_csv()

    # In ra một số dòng để bạn check thứ tự row_idx và book_id
    print("\\n[Thành công] Nội dung của Index CSV:")
    display(id_map_df.head(10))
    print(f"\\nTổng cộng số dòng CSV tạo ra là {len(id_map_df)}")

except Exception as e:
    print(f"Lỗi xảy ra trong quá trình sinh CSV độc lập: {e}")
```
